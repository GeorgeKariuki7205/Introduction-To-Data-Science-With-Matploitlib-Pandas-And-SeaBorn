# **CREATING** **VISUALISATIONS.**

```
# We are now grouping the new data set, it is made by using the doube square brackets.
grouped = cd.groupby('province')[['no_grass','no_tiles','no_concrete']].sum()
```

```
#showing the new data set.
grouped
```

```
# grouping and using the ascending order functionality
grouped.sort_values(by = 'no_grass', ascending = False)
```

```
grouped.sort_values(by = 'no_grass', ascending = True)
```

```
# Importing our plotting libraries
import seaborn as sns
import matplotlib.pyplot as plt
```

```
# plotting the data that has been said 
grouped.plot(kind = 'bar', figsize = (8,5), logy = True)
plt.title('The graph to show the roofing in the various provinces.', color = 'red',fontsize = 13)
plt.xlabel('Province', color = 'darkblue',
            fontsize = 13)
plt.ylabel('Number of roofing materials.', color = 'darkblue',
            fontsize = 13)
plt.legend(['Grass','Tiles','Concrete'])
plt.xticks(rotation = 45)

plt.show()
```

```
#box plots for getting the data...

plt.figure(figsize = (10,8))
sns.boxplot(cd.households_corrugated_iron_sheets,color ='darkgreen')
plt.title('This is the box plot showing the ouliers')
plt.xlabel('Corrugated Iron sheets')
plt.show()

# cd.columns
```
